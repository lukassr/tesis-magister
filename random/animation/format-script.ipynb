```python
import pandas as pd
import re
import json

tsv_file = 'grid-rm.tsv'
df = pd.read_csv(tsv_file,sep='\t')
df.head()
```

```python
## Output for the animation geo2rect
with open('output.json','w') as f:
    f.write('{')
    for k,v in df.T.to_dict().items():
        f.write("'"+str(v["region"])+"':"+"{x:"+str(v["gridX"])+",y:"+str(v["gridY"])+"},\n")
    f.write("}")
```

```python
## Output CSV file
pd.read_csv(tsv_file,sep='\t').to_csv("grid-rm.csv",index=True)
```
